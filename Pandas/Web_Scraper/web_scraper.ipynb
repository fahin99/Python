```python
import pandas as pd
```

### 1. Read a .csv from an url w Pandas
#### Target site: https://www.football-data.co.uk/data.php

```python
# reading the csv file
url='https://www.football-data.co.uk/mmz4281/2526/E0.csv'
df_epl=pd.read_csv(url) 
```

```python
# FTHG & FTAG terms are confusing, while they just mean 
# final time home goals and aways goals
df_epl.rename(columns={'FTHG':'home_goals', 
                          'FTAG':'away_goals'}, inplace=True)
df_epl
```

### 2. Working with multiple .csv files

```python
# now we read multiple csvs and add them to a list
# for that we need a root url
root="https://www.football-data.co.uk/mmz4281/"

leagues=['E0', 'E1', 'E2']
```

```python
# now we loop thru the legues array and read the corresponding csvs
frames=[]
for league in leagues:
    df=pd.read_csv(root+'2526/'+league+'.csv')
    frames.append(df) 
```

```python
#check the length of "frames"
len(frames)
```

```python
frames[1]
```

```python
for df in frames:
    df.rename(columns={'FTHG':'home_goals', 
                          'FTAG':'away_goals'}, inplace=True)

frames[1]
```

#### Multiple seasons

```python
frames=[]
for league in leagues:
    for season in range(20,26):
        df=pd.read_csv(root+str(season)+str(season+1)+'/'+league+'.csv', encoding='unicode_escape')
        df['season']=season
        frames.append(df) 
```

```python
len(frames)
```

### 3. Organize all the data in a  dictionary

```python
dict_countries={'La Liga':'SP1', 'Spanish Segunda Div':'SP2',
                'Bundesliga':'D1',
                'EPL':'E0', 'Eng League 1':'E1'}
```

```python
dict_historical_data={}

# looping thru keys and creating list of leagues

frames=[]

for league in dict_countries:
    for season in range(20,26):
        df=pd.read_csv(root+str(season)+str(season+1)+'/'+dict_countries[league]+'.csv', encoding='unicode_escape')
        df.insert(1, 'Season', season)
        frames.append(df)
    # here we create frame for each league and add it to the dictionary with the league name as key
    df_net=pd.concat(frames)
    dict_historical_data[league]=df_net 
```

### Create column based on 1 conditions: np.where()

```python
import numpy as np

# create an array based on AvgCAHH greater than 1.92

np.where(df_epl['AvgCAHH']>=1.92, 'High', 'Low')
```

```python
# set it to a new column
df_epl['AvgCAHH_tier']=np.where(df_epl['AvgCAHH']>=1.92, 'High', 'Low')

df_epl
```

```python
df_epl.value_counts('AvgCAHH_tier')
```

### Creating column from more than 2 choices: np.select()

```python
conditions=[
    df_epl['AvgCAHH']>=2,
    (df_epl['AvgCAHH']<2) & (df_epl['AvgCAHH']>=1.8),
    df_epl['AvgCAHH']<1.8
]

values= ['High', 'Mid', 'Low']
```

```python
df_epl['AvgCAHH_score']= np.select(conditions, values, default='Unknown')
```
